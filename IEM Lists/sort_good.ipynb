```python
import re

def read_file(filename):
    with open(filename, 'r') as f:
        content = f.read().strip()
    return content
```

```python
def parse_content(content):
    lines = content.split('\n')
    parsed_data = {}
    processed_count = 0
    unprocessed_count = 0
    keys = set()
    
    for line in lines:
        if line.strip():
            match = re.match(r"'(.+)' : \[([^\]]+)\]", line)
            if match:
                key, values = match.groups()
                if key in keys:
                    print(f"Duplicate key: {key}")
                else:
                    keys.add(key)
                    parsed_data[key] = [val.strip() for val in values.split(',')]
                    processed_count += 1
            else:
                print(f"Line not captured: {line}")
                unprocessed_count += 1
                
    print(f"Processed lines: {processed_count}")
    print(f"Unprocessed lines: {unprocessed_count}")
    return parsed_data
```

```python
def sort_data(data):
    sorted_items = sorted(data.items(), key=lambda x: x[0].lower())
    sorted_data = {}
    for key, values in sorted_items:
        if key not in sorted_data:
            sorted_data[key] = values
        else:
            print(f"Duplicate key found: {key}")
            new_key = f"{key} (duplicate)"
            while new_key in sorted_data:
                new_key += " (duplicate)"
            sorted_data[new_key] = values
    return sorted_data
```

```python
def format_sorted_data(sorted_data):
    formatted_data = []
    for key, values in sorted_data.items():
        formatted_values = ', '.join(values)
        formatted_line = f"'{key}' : [{formatted_values}]"
        formatted_data.append(formatted_line)
    return '\n'.join(formatted_data)
```

```python
def write_sorted_data(filename, sorted_data):
    with open(filename, 'w') as f:
        f.write(sorted_data)
```

```python
input_filename = 'sorted_good list.txt'
output_filename = 'output.txt'

content = read_file(input_filename)
data = parse_content(content)
sorted_data = sort_data(data)
formatted_sorted_data = format_sorted_data(sorted_data)
write_sorted_data(output_filename, formatted_sorted_data)
```
