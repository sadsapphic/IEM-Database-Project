```python
import re
from collections import defaultdict

def extract_iems(filename):
    with open(filename, 'r') as file:
        data = file.read()
        
    iem_pattern = re.compile(r'(\w+-\d+|[A-Za-z]+ \d+|[A-Za-z]+ [A-Za-z]+ [A-Za-z]+|[A-Za-z]+-[A-Za-z]+)')
    iems = iem_pattern.findall(data)
    
    iem_dict = defaultdict(list)
    for iem in iems:
        main_name = iem.split(' ')[0]  # Get the main name by extracting the first word
        if iem not in iem_dict[main_name]:
            iem_dict[main_name].append(iem)  # Add the IEM to the list of variations

    # Sort the dictionary alphabetically by key and the variations list for each key
    sorted_iem_dict = {k: sorted(v) for k, v in sorted(iem_dict.items())}

    return sorted_iem_dict

filename = "list.txt"  # Replace with your .txt file name
result = extract_iems(filename)
print(result)
```
